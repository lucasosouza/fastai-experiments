# Cifar-10 on VGG

### Load libraries

```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
import fastai
from fastai.vision import *
from fastai.metrics import error_rate
```

```python
# check for cuda
import torch.cuda
torch.cuda.is_available()
```

### Load data

```python
path = untar_data(URLs.CIFAR)
```

```python
tfms = get_transforms(do_flip=False)
```

```python
path.ls()
```

```python
(path/'train').ls()
```

```python
(path/'test').ls()
```

```python
data = (ImageList.from_folder(path) #Where to find the data? -> in path and its subfolders
        .split_by_folder(train='train', valid='test')              #How to split in train/valid? -> use the folders
        .label_from_folder()            #How to label? -> depending on the folder of the filenames
        .transform(tfms, size=32)       #Data augmentation? -> use tfms with a size of 64
        .databunch()) 
```

```python
data.show_batch(3, figsize=(6,6), hide_axis=False)
```

### Train model

```python
# dir(models)
```

```python
learn = cnn_learner(data, models.vgg19_bn, metrics=error_rate, pretrained=False)
```

```python
# learn.model
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

### Improving

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-5,1e-3))
```

```python
learn.save('stage-2')
```
